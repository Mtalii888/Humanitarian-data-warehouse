```python
import pandas
```

```python
cash_data = pandas.read_csv(r"C:\Users\ELITEX21012G2\Desktop\REFUGEES PROJECT\data\raw\cash_assistance_raw.csv")
cash_df = pandas.DataFrame(cash_data)
cash_df.head()
```

```python
cash_df.info()
```

```python
cash_df = cash_df.dropna()
```

```python
cash_df["payment_date"] = pandas.to_datetime(cash_df["payment_date"], format='mixed', errors='coerce')
```

```python
cash_df.head()
```

```python
cash_df["country"].unique()
```

```python
cash_df["country"] = cash_df["country"].str.strip()
```

```python
cash_df["country"] = cash_df["country"].str.title()
```

```python
cash_df["currency"].unique()
```

```python
cash_df["payment_method"].unique()
```

```python
cash_df.duplicated().sum()
```

```python
cash_df = cash_df.drop_duplicates()
```

```python
cash_df.describe().T
```

```python
cash_df["amount"] = cash_df["amount"].abs()
```

```python
cash_df.rename(columns={"country":"country_id"}, inplace=True)
```

```python
countries_lookup = {"Kenya" : 1,
                    "Uganda" : 2,
                    "South Sudan" : 3,
                    "Ethiopia" : 4,
                    "Somalia" : 5,
                    "Dr Congo" : 6,
                    "Rwanda" : 7,
                    "Burundi" : 8,
                    "Tanzania" : 9,
                    "Sudan" : 10,}
```

```python
cash_df["country_id"] = cash_df["country_id"].map(countries_lookup)
```
