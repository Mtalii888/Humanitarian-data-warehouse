```python
countries_df = countries_df.drop_duplicates()
```

```python
countries_df
```

```python
countries_df["country_name"].unique()
```

```python
countries_df["country_name"] = countries_df["country_name"].str.title()
countries_df["country_name"] = countries_df["country_name"].str.strip()
```

```python
countries_lookup = {"Kenya" : 1,
                    "Uganda" : 2,
                    "South Sudan" : 3,
                    "Ethiopia" : 4,
                    "Somalia" : 5,
                    "Dr Congo" : 6,
                    "Rwanda" : 7,
                    "Burundi" : 8,
                    "Tanzania" : 9,
                    "Sudan" : 10,}
```

```python
countries_df["country_id"] = countries_df["country_name"].map(countries_lookup)
```
