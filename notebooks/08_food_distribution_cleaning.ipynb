```python
import pandas
```

```python
food_data = pandas.read_csv(r"C:\Users\ELITEX21012G2\Desktop\REFUGEES PROJECT\data\raw\food_distribution_raw.csv")
food_df = pandas.DataFrame(food_data)
food_df.head(10)
```

```python
food_df.isnull().sum()
```

```python
food_df = food_df.dropna()
```

```python
food_df[food_df.duplicated()]
```

```python
food_df = food_df.drop_duplicates()
```

```python
food_df.info()
```

```python
food_df["distribution_date"] = pandas.to_datetime(food_df["distribution_date"],format='mixed', errors='coerce')
```

```python
food_df["distribution_date"].isnull().sum()
```

```python
food_df["food_item"].unique()
```

```python
food_df["country"].unique()
```

```python
food_df["country"] =food_df["country"].str.strip()
```

```python
food_df["country"] =food_df["country"].str.title()
```

```python
food_df.describe().T
```

```python
food_df["quantity"] = food_df["quantity"].abs()
food_df.head(10)
```

```python
food_df =  food_df.rename(columns={"country":"country_id"})
```

```python
countries_lookup = {"Kenya" : 1,
                    "Uganda" : 2,
                    "South Sudan" : 3,
                    "Ethiopia" : 4,
                    "Somalia" : 5,
                    "Dr Congo" : 6,
                    "Rwanda" : 7,
                    "Burundi" : 8,
                    "Tanzania" : 9,
                    "Sudan" : 10,}
```

```python
food_df["country_id"] = food_df["country_id"].map(countries_lookup)
```
